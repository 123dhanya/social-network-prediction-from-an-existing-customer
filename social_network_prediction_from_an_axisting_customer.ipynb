```python
# how to drop the coulmn for the dataset
data.drop('Gender',axis=1,inplace=True)
data.drop('User ID',axis=1,inplace=True)

data.tail(5) # display the last 5 value
```

#split input & output

```python
x=data.iloc[:,:-1].values # :-1 leave the last column and take the another columns
x

```

```python
y=data.iloc[:,-1].values # -1 take only last columns
y
```

#splitting dataset to train & test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

#feature scaling [ we scale our data to make all the feature contribute equally to the result]

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform (x_train) # fit calculating the mean & variance of each of the feature  present in our data. 
x_test = sc.transform (x_test) # transforming all the feature using the respective mean & variance
print(x_train)
```

```python
# import logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)) 
```

#**accuracy score**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print("accuracy of the model:{0}%".format(accuracy_score(y_test, y_pred)*100))
cm=confusion_matrix(y_test,y_pred)
print("confusion_matrix:",cm)
```
